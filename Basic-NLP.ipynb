```python
uniqueemo=df['emotion'].unique()
uniqueemo
```

```python
emo_num={}
i=0
for emo in uniqueemo:
  emo_num[emo]=i
  i+=1

df['emotion']=df['emotion'].map(emo_num)
df
```

```python
df['text'] = df['text'].apply(lambda x: x.lower())
df.head()
```

TO remove punction we also have function like string.punctuation


```python
import string
def remove_punc(text):
  return text.translate(str.maketrans('','',string.punctuation))
  #return text.translate(str.maketrans('', '',string.punctuation))
```

```python
df['text']=df['text'].apply(remove_punc)
```

```python
def remove_num(text):
  new=""
  for i in text:
    if not i.isdigit():
      new= new +i
  return new

df['text']=df['text'].apply(remove_num)
#print(new)
```

```python
def remove_emojis(text):
  new = " "
  for i in text:
    if i.isascii():
      new += i
  return new.strip()

df['text'] = df['text'].apply(remove_emojis)
```

```python
stop_words=set(stopwords.words('english'))
len(stop_words)
```

tokenization-convert string to single token
Not to use split is that it cant diff punctuation marks


```python
def remove(txt):
  words=word_tokenize(txt)
  return words
```

```python
df['tokenized_text'] = df['text'].apply(remove)
df.head()
```

```python
def remove_stopwords(text_tokens):
  return [word for word in text_tokens if word not in stop_words]

df['cleaned_text'] = df['tokenized_text'].apply(remove_stopwords)
df.head()
```
